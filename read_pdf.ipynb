```python
urlpdf="http://www.cbr.ru/collection/collection/file/7783/bbsq0402r.pdf"
webbrowser.open(urlpdf)
pages=["5-6", "12-14"]
```

```python
## convert PDF tables and save them in CSV, XLSX files,
##  and in a dictionary of DataFrames >>
#
df = {}
wb = Workbook()
for i in range(1,len(pages)+1):
  df[i] = pd.DataFrame()
  tmpFileName = 'tmp'+str(i)+'.csv'
  j = i-1
  print('\ndf[{0}] <-'.format(i),tmpFileName+', pages',
        pages[j], " >>>")
  # convert PDF into CSV file >>
  tabula.convert_into(urlpdf, tmpFileName, output_format="csv",
                    pages=pages[j],
                    java_options="-Dfile.encoding=UTF8")
  # convert CSV into XLSX file
  wb.create_sheet(pages[j], j) # <- creates a sheet
  wb._active_sheet_index = j   # <- sheet number to write in
  ws = wb.active
  with open(tmpFileName, 'r', encoding="utf8") as f:
    for row in csv.reader(f):
        ws.append(row)
  # convert to DataFrame >>
  df[i] = pd.read_csv(tmpFileName)
  print('\n', df[i])
wb.save('tmp.xlsx')
wb.close()
print('\n>>> Data saved in\n   ', os.getcwd()+r'\tmp.xlsx')
```

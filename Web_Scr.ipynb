```python
#
## Regular Expressions
#

page = urlopen(url)
html = page.read().decode("utf-8")

print('\n', html)
```

```python
# .*? is everything the shortest possible >>
pattern = "<title.*?>.*?</title.*?>"
match_results = re.search(pattern, html, re.IGNORECASE)

print('\n', match_results)
```

```python
title = match_results.group()

print('\n', title)
```

```python
# remove HTML tags >>
title = re.sub("<.*?>", "", title)

print('\n', title)
```

```python
import webbrowser
import re
import mechanicalsoup
import pandas as pd
import time
import os
```

```python
## Список кредитных организаций (Банк России)
#
url = "https://www.cbr.ru/banking_sector/credit/FullCoList/"
webbrowser.open(url)
```

```python
#
## Create Browser Object
#
browser = mechanicalsoup.Browser()
page = browser.get(url)
```

```python
# get URL for each link of the page
links = page.soup.select("a")
```

```python
## save banks' URLs in DataFrame >>
#
base_url = "https://www.cbr.ru"
df=pd.DataFrame(columns=['name', 'url'])
i=0
for link in links:
    address = link["href"]
    m = re.match(r"/banking_sector/", address)
    if m:
        text = link.text
        print(f"{i} > {text}: {address}")
        df.loc[i]=[text] + [base_url + address]
        i += 1
```

```python
otz = page.soup.select("td")
display(otz)
```

```python
## find banks' License status >>
j=1
for i in otz:
  t = i.text
#  print(f"{i} -> {t}")
  m = re.match(str(j), t)
  if m:
   j += 1
   t2=i.find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").text
   if t2!='\n':
    print(f"{t} -> {t2}", end="  ")
```

```python
## save banks' License status in DataFrame >>
#
dfst=pd.DataFrame(columns=['st'])
j=1
for i in otz:
  t = i.text
  m = re.match(str(j), t)
  if m:
   j += 1
   t2=i.find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").find_next("td").text
   if t2!='\n':
#    print(f"{t}: >>> {t2} <<<")
    print(f"{t} -> {t2}", end="  ")
    dfst.loc[j]=[t2]

print('\n', dfst)
```

```python
# U may find some errors >>>
print( '\n', dfst[(dfst.st != "ОТЗ") &
                  (dfst.st != "АНН") &
                  (dfst.st != "РЕГ")] )
# the same with a nice Python code >>
print( '\n', dfst[ ~dfst.st.isin(["ОТЗ","АНН","РЕГ"]) ] )
```

```python
# U may keep good rows in a DataFrame >>>
print('\n', dfst[(dfst.st=="ОТЗ") |
                 (dfst.st=="АНН") |
                 (dfst.st=="РЕГ")].value_counts() )
# keep using a nice Python code >>
df_tmp = dfst[ dfst.st.isin(["ОТЗ","АНН","РЕГ"]) ]
# just for info >>
print('\n>>> We have >>>\n', df_tmp.st.value_counts())
```

```python
## merge in one dataframe >>>
#
df_st=pd.concat([df, df_tmp], axis=1)

print('\n', df_st)
```
